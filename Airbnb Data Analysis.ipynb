```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Airbnb_data.csv.csv")
```

```python
df.head()
```

```python
df.columns
```

# checking missing values

```python
print(df.isnull().sum())
```

```python
df.info()
```

# Handling missing values

```python
df['last review']=pd.to_datetime(df['last review'],errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' :0,'last review':df['last review'].min()},inplace=True)
```

```python
df.dropna(subset = ['NAME','host name'], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df=df.drop(columns=['license','house_rules'],errors='ignore')
```

```python
df['price']=df['price'].replace('[\$]','',regex=True).astype(float)
df['service fee']=df['service fee'].replace('[\$]','',regex=True).astype(float)
```

```python
df.head()
```

```python
#remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

# Visualization

# PROBLEM STATEMENTS
# WHAT IS THE DISTRIBUTION OF LISTING PRICES?

```python
#PROBLEM STATEMENTS
#1.WHAT IS THE DISTRIBUTION OF LISTING PRICES?
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=50,kde=True,color='red')
plt.title("Distribution of listing price")
plt.xlabel('Price $')
plt.ylabel('Frequency')
plt.show()
```

# HOW ARE DIFFERENT ROOM TYPES DISTRIBUTED?

```python
plt.figure(figsize=(8,5))
sns.countplot(x='room type',data=df,color="yellow")
plt.title('Room Type Distribution')
plt.xlabel('room type')
plt.ylabel('count')
plt.show()
```

# HOW ARE LISTINGS DISTRIBUTED ACROSS DIFFERENT NEIGHBORHOODS?

```python
plt.figure(figsize=(12,8))
sns.countplot(y='neighbourhood group',data=df,color="green",order=df['neighbourhood group'].value_counts().index)
plt.title('Number of listings by Neighborhood Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')
plt.show()
```

# WHAT IS THE RELATIONSHIP BETWEEN PRICE AND ROOM TYPE?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type',y='price',hue='room type',data=df,palette='Set1')
plt.title('Price vs Room type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show
```

# HOW HAS THE NUMBER OF REVIEWS CHANGED OVER TIME?

```python
df['last review']=pd.to_datetime(df['last review'])
reviews_over_time=df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number Of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number Of Reviews')
plt.show()
```

# Correlation Heatmap for numerical features

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

# Top 10 Hosts with Most Listings

```python
top_hosts = df['host name'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_hosts.values, y=top_hosts.index, palette='viridis')
plt.title("Top 10 Hosts with Most Listings")
plt.xlabel("Number of Listings")
plt.ylabel("Host Name")
plt.show()
```
